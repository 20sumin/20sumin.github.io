#### [문제5]

다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.

1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)


```python
def addall(list):
    result = 0
    for index in range(len(list)):
        result += list[index]
    return result

print(addall([1, 2, 3]))
```

- for문을 리스트의 크기만큼 반복하여 리스트 요소의 합을 구한 뒤 반환한다.

```python
def addall(list):
    result = [x for x in list]
    return sum(result)

print(addall([1, 2, 3]))
```

- 리스트 내포를 통해 리스트의 요소를 받아와서 sum()함수를 통해 리스트 요수의 합을 구한 뒤 반환한다.

2) addallodd(list) 함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.)

```python
def addallodd(list):
    result = [x for x in list if x%2 != 0]
    return sum(result)

print(addallodd([1, 2, 3, 4, 5]))
```

- 리스트 내포를 통해 리스트의 요소 중 홀수 만 받아 온 후, sum()함수를 통해 홀수의 합을 구해 반환한다.

```python
def addallodd(list):
    result = 0
    for index in range(len(list)):
        if list[index]%2 != 0:
            result += list[index]
    return result

print(addallodd([1, 2, 3, 4, 5]))
```

- 반복문을 리스트의 크기만큼 반목하여 리스트가 홀수일 때 그 요소를 더해 최종적으로 홀수의 합을 구해 반환한다.

#### [문제 10]

다음 문제를 정확히 읽고 적합한 코딩을 제시하시오.

- 1번 그룹, 2번 그룹, ... n번 그룹으로 번호가 매겨진 총 n개의 그룹이 존재하고, ith 그룹에는 임의의 팀원 수 ai 명이 있다.하나의 버스 정류장이 있고, 버스는 오직 한 대만 운영한다. 버스는 버스 정류장에 도착하고 모든 그룹은 버스를 통해 여행하길 원한다.하지만, 같은 그룹에 속한 팀원들은 분리되기를 원치 않는다. 그래서 그룹의 모든 팀원이 탈 수 있을 때 에만 그 그룹은 버스에 탑승한다.또한, 이들은 그룹의 번호순으로 여행을 하길 원한다. 예를 들어, 3번 그룹은 1번 그룹과 2번 그룹이 여행을 한 뒤에 버스를 탈 수 있거나, 1번 그룹과 2번 그룹이 이미 버스에 탑승한 상태에 있을 때 동시에 버스를 탈 수 있다.마지막으로, 버스는 빈 좌석이 없을 때 에만 여행을 떠난다.

- 이 문제에서 각 그룹의 팀원 수가 리스트로 주어질 때, 가능한 모든 버스의 사이즈 x(x명의 사람들이 동시에 탈 수 있음)를 반환하는 함수 solve(a) 를 작성하시오.
- [주의] 반드시 계산 복잡도를 최대한 낮추세요. 아래 예시 3번을 실행했을 때 결과가 바로 나타날 수 있도록!!

```python
def solve(a):
    result = []
    total = sum(a)
    people = 0
    for index in range(len(a)):
        people += a[index]
        tmp = 0
        if total % people == 0:
            i = index + 1
            while i < len(a):
                tmp += a[i]
                if tmp == people:
                    tmp = 0
                    if i == len(a)-1:
                        result.append(people) 
                        break;
                elif tmp > people:
                    break
                i += 1
    result.append(total)
    return result

print(solve([1, 2, 1, 1, 1, 2, 1, 3]))
```

- 인덱스에 해당하는 요소를 를 순서대로 for이 반복할 때만다 people에 더해진다.
- people의 값을 비교하여 요소 전체 합과 나누어 떨어지는 경우 while문을 통해 탑승 가능한 숫자이지 판별한다.

```python
import re #정규 표현식
import requests

req1=requests.get('https://www.naver.com/')
s1=req1.text

#re.sub(pattern, repl, string[, count])
#string에서 pattern과 일치하는 부분에 대해 repl로 교체해서 결과가 문자열을 반환
#.는 글자가 없을 수도 있다는 것을 의미
#*는 문자가 0회이상 반복된다는 것을 의미
#여러 줄을 대상으로 찾아야 하기 때문에 0으로 명시한다.
#re.l은 대소문자를 구분하지 않겠다는 것을 의미
#re.S는 여러줄에 걸쳐서 정규표현식을 찾으라는 의미
s1=re.sub('<script.*?>.*?</script>','',s1,0,re.I|re.S)
#<script로 시작하고 그 위에 어떤 글자가 없을 수도 있고 여러개 있을 수도 있으며
#그 위에는 >가 나오고 그 위에 다시 어떤 글자가 없을 수도 있고 여러개 있을 수고 있으며
#마지막은 </script>로 끝난다.

s1=re.sub('<style>.*?</style>','',s1,0,re.I|re.S)
s1 = re.sub('<.+?>', '', s1, 0, re.I|re.S )
s1 = re.sub('{.+?}', '', s1, 0, re.I|re.S)
s1 = re.sub('\n|\t', '', s1)
s1_list = s1.split()
for index in range(len(s1_list)): #구두 문자를 제거한다.
    s1_list[index] = ''.join(char for char in s1_list[index] if char.isalnum())
s1_list = [string for string in s1_list if string != '']

s1_dict = {}
s1_list2 = list(set(s1_list))
for string in s1_list2:  
    s1_dict[string] = s1_list.count(string)
    #count() 함수를 통해 키값과 키가 몇개있는지를 value 값으로
    #s1_dict에 삽입한다.

print(s1_dict)
```
