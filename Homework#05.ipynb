#### [문제 2]
다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.
- 다형성이란, 상속 관계에 있는 클래스들의 객체들이 같은 멤버 함수를 호출하지만 각각 코드가 실행되도록 하는 것이다.


```python
from math import*

class Circle: #원
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return pi * (self.radius ** 2) #원의 넓이 : pi*r^2
    
class Sphere(Circle): #구
    def __init__(self, radius):
        super(Sphere, self).__init__(radius) #부모 클래스 생성자 호출
                                             #bound 메소드 호출
    def area(self): #area() 함수 오버라이딩
        return 4 * pi * (self.radius ** 2) #구의 겉넓이 : 4pi * r^2
    
circle = Circle(2)
sphere = Sphere(2)

print(circle.area())
print(sphere.area())
```

#### [문제 3]
다음 각 요구사항을 만족시키는 Counter 클래스를 코딩하시오.
- 생성자에 count 값과 step값을 인자로 받을 수 있다.
    - count : Counter 객체가 지니는 초기 정수 값
    - step : Counter 객체의 count가 증가되는 증분(default = 1)
- Counter의 객채를 출력해주는 __ __str__ __() 메소드를 구현
- step에 주어진 증분만큼 count를 증가시키는 incr() 메소드 구현
- Counter 클래스 내에 관련 메소드를 추가하여 객체를 직접 호출할 수 있도록 한다.
- (+, -) 연산을 수행하는 메소드 추가
- (>, < ==) 연산을 수행하는 메소드 추가

```python
class Counter:
    #요구사항 1.
    def __init__(self, count, step = 1):
        self.count = count
        self.step = step
        
    #요구사항 2.
    def __str__(self):
        return "[Count(step: {})] {}".format(self.step, self.count)
    
    #요구사항 3.
    def incr(self):
        self.count += self.step
       
    #요구사항 4.
    def __call__(self):
        self.incr()
    
    #요구사항 5.
    def __add__(self, other):
        self.count += other
        return self
    
    def __sub__(self, other):
        self.count -= other
        return self
    
    #요구사항 6.
    def __gt__(self, other):
        return self.count > other
    
    def __eq__(self, other):
        return self.count == other
    
    def __lt__(self, other):
        return self.count < other 

#요구사항 1, 2.
a = Counter(10)
b = Counter(10, 2)
print(a)
print(b)

#요구사항 3.
a.incr()
b.incr()
print(a)
print(b)

#요구사항 4.
a()
b()
print(a)
print(b)

#요구사항 5.
a = a + 5
b = b - 5
print(a)
print(b)

#요구사항 6.
print(a > 10)
print(b > 10)
print(a < 10)
print(b < 10)
print(a == 17)
print(b != 9)
```

- 요구사항 1.
    - count와 step을 인자로 받아와 멤버 변수를 초기화한다.
    - step의 default 값은 1로 정의한다.
- 요구사항 2.
    - format() 함수를 통해 요구사항과 같은 형식으로 문자열을 반환한다.
- 요구사항 3.
    - 객체의 count에 step의 값 만큼 더한다.
- 요구사항 4.
    - 객체를 직접 호출할 때 호출되는 매직 메소드 __ __call__ __()을 재정의하여, incr() 함수를 호출한다.
- 요구사항 5.
    - 덧셈과 뺄셈에 해당하는 매직 메소드를 재정의하여 객체의 count 값에 정수가 더해지거나 빠지도록 한다.
- 요구사항 6.

#### [문제 4]
내장 자료형 list를 서브클래싱 하는 MySet 클래스의 정의 내용을 보고 __ __init__ __(), __ __str__ __(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 다른 사람에게 가르친다고 생각하며 설명해보시오.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

- __ __init__ __(self, l)
    - __ __init__ __()은 생성자로 클래스 객체가 생성될 때, 호출되는 메소드이다.
    - MySet의 생성자는 리스트 l을 인자로 받으며, 객체에 받은 리스트 l의 요소를 반복문을 통해 추가한다. 그리고 클래스 함수 eliminate_duplicate()에 인자를 self로 넣어 호출한다.
    - self 객체는 list를 서브클래싱한 클래스의 객체이므로, 리스트를 이용할 때 사용하는 함수를 사용할 수 있다.
- __ __str__ __(self)
    - __ __str__ __()은 print 예약어나 str() 내장함수 호출에 대응되는 메소드로 "MySet: { MySet 객체의 요소들 }" 형식으로 문자열을 반환한다. 이때, 반복문을 통해 요소들은 콤마(,)로 구분한다.
- elimicate_duplicate(l)
    - 인자로 리스트를 받는 클래스 함수이다. 새로운 리스트 s를 생성하고 리스트 l의 요소가 이미 리스트 s에 있다면 s에 추가하지 않도록 하여, 리스트 l에서 중복되는 요소가 없도록 리스트 s에 추가한다.
    - 리스트 l을 슬라이싱을 통해 초기화하고 리스트 s의 요소를 다시 리스트 l에 추가한다.

#### [문제 5]
MySet 클래스에 메소드를 추가하여 요구사항을 모두 만족시키는 코딩을 제시하시오.
- 요구사항 1.
    - | 연산으로 두 집합의 합집합을 반환한다.
- 요구사항 2.
    - & 연산으로 두 집합의 교집합을 반환한다.
- 요구사항 3.    
    - -연산으로 두 집합의 차집합을 반환한다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result
    
    #요구사항 1.
    def __or__(self, other): #연산자 |의 매직 메소드 재정의
        s = MySet([])
        for e in (self + other):
            if e not in s: #s에 없는 요소만 s에 추가하여 합집합을 완성한다.
                s.append(e)
        return s
    
    #요구사항 2.
    def __and__(self, other): #연산자 &의 매직 메소드 재정의
        s = MySet([])
        for e_s in self:
            for e_o in other: #self와 other에 모두 있는 요소만 s에 추가하여 교집합을 완성한다.
                if e_s == e_o: 
                    s.append(e_o)
        return s
    
    #요구사항 3.
    def __sub__(self, other): #연산자 -의 매직 메소드 재정의
        s = MySet([])
        for e_s in self:
            s.append(e_s)
            for e_o in other:  #self의 요소를 가지고 있는 s에서
                if e_s == e_o: #self와 other에 모두 있는 요소를 s에서 제거하여 차집합을 완성한다.
                    s.remove(e_o)
        return s

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    
    #요구사항 1.
    u = s | t
    print(u)
    
    #요구사항 2.
    u = s & t
    print(u)
    
    #요구사항 3.
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```

#### [Incremental Project 문제]
1. Assignment3 & Assignment4의 Incremental Project 코드를 다시 확장/변형하여 다음과 같은 조견을 만족하도록 구현하시오.
> 1) 새로운 클래스 SearchEngine를 정의하시오

```python
import re #정규 표현식
import requests
import string

class SearchEngine:
    #2번
    def __init__(self, *args):
        self.url_list = []
        self.url_dict = {}
        for url in args:
            self.url_list.append(url)
    
    #3번
    def addUrl(self, url):
        if url not in self.url_list:
            self.url_list.append(url)
    
    #4번
    def removeUrl(self, url):
        if url in self.url_list:
            self.url_list.remove(url)
    
    #5번
    def listUrls(self):
        for url in self.url_list:
            print(url)
     
    #6번
    def getWordsFrequency(self):
        sources_list = []
        for url in self.url_list:
            req = requests.get(url)
            source = req.text

            body = re.search('<body.*/body>', source, re.I|re.S)
            source = body.group()

            source = re.sub('<script.*?>.*?</script>','',source,0,re.I|re.S)
            source = re.sub('<style>.*?</style>','',source,0,re.I|re.S)
            source = re.sub('<.+?>', '', source, 0, re.I|re.S )
            source = re.sub('{.+?}', '', source, 0, re.I|re.S)
            source = re.sub('http.+"', '', source, 0, re.I|re.S)
            source = re.sub('★|©|@|\n|\t', ' ', source)
            source = re.sub('middot', ' ', source)
            source = source.translate(str.maketrans('', '', string.punctuation))

            source_list = source.split()
            for index in range(len(source_list)):
                source_list[index] = ''.join(char for char in source_list[index] if char.isalnum())
            source_list = [string for string in source_list if string != '']
            sources_list += source_list

        source_dict = {}
        source_list2 = list(set(sources_list))
        for s in source_list2:
            source_dict[s] = sources_list.count(s)
    
        with open('Korean.txt', 'r', encoding = 'utf-8') as f:
            korean = f.read()
        with open('English.txt', 'r', encoding = 'utf-8') as f:
            english = f.read()

        k_list = korean.split("\n")
        e_list = english.split("\n")

        for word, cnt in source_dict.items():
            if word not in (k_list + e_list):
                self.url_dict[word] = cnt
        
        return self.url_dict
    
    #7번
    def getMaxFrequencyWords(self):    
        if not self.url_dict:
            return None
        else:
            rank = sorted(self.url_dict.items(), key = lambda x : x[1], reverse = True)
            for index in range(len(rank)):
                if rank[index][1] != rank[index + 1][1]:
                    break
            
            rank = dict(rank[:index + 1])
            return rank
```

```python
class SearchEngineWithOrderedWebWords(SearchEngine): 
    def getWordsFrequency(self, reverse = False):
        if reverse: reverse = False
        else: reverse = True
        
        sources_list = []
        for url in self.url_list:
            req = requests.get(url)
            source = req.text

            body = re.search('<body.*/body>', source, re.I|re.S)
            source = body.group()

            source = re.sub('<script.*?>.*?</script>','',source,0,re.I|re.S)
            source = re.sub('<style>.*?</style>','',source,0,re.I|re.S)
            source = re.sub('<.+?>', '', source, 0, re.I|re.S )
            source = re.sub('{.+?}', '', source, 0, re.I|re.S)
            source = re.sub('http.+"', '', source, 0, re.I|re.S)
            source = re.sub('★|©|@|\n|\t', ' ', source)
            source = re.sub('middot', ' ', source)
            source = source.translate(str.maketrans('', '', string.punctuation))

            source_list = source.split()
            for index in range(len(source_list)):
                source_list[index] = ''.join(char for char in source_list[index] if char.isalnum())
            source_list = [string for string in source_list if string != '']
            sources_list += source_list

        source_dict = {}
        source_list2 = list(set(sources_list))
        for s in source_list2:
            source_dict[s] = sources_list.count(s)
    
        with open('Korean.txt', 'r', encoding = 'utf-8') as f:
            korean = f.read()
        with open('English.txt', 'r', encoding = 'utf-8') as f:
            english = f.read()

        k_list = korean.split("\n")
        e_list = english.split("\n")

        for word, cnt in source_dict.items():
            if word not in (k_list + e_list):
                self.url_dict[word] = cnt
        
        result = sorted(self.url_dict.items(), key = lambda x : x[1], reverse = reverse)
        print(result)
        
    def __iter__(self):
        return iter(list(self.url_dict.items()))
```

- 부모 클래스의 getWordsFrequency() 함수의 코드를 통해 얻은 사전을 sorted()함수를 통해 리스트를 새로운 변수 result에 받는다. 이때, sorted() 함수에서 reverse의 값에 인자로 받은 reverse를 대입하여 False인 경우에는 내림차순이고 True인 경우에는 오름차순이 된다.
